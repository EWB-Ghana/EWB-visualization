```python
import os
import folium
```

```python
from folium import FeatureGroup, LayerControl, Map, Marker

m = Map(
    location=[45.372, -121.6972],
    zoom_start=12,
    tiles='Stamen Terrain'
)

#location = []

group_boreholes = FeatureGroup(name='Boreholes')
Marker(location=[45.3288, -121.6625],
       popup='JHS Borehole').add_to(group_boreholes)

Marker(location=[45.3311, -121.7113],
       popup='Borezy Well').add_to(group_boreholes)

group_houses = FeatureGroup(name='Houses')

Marker(location=[45.370, -121.80], popup='House A').add_to(group_houses)
Marker(location=[45.360, -121.6625], popup='House B').add_to(group_houses)


group_boreholes.add_to(m)
group_houses.add_to(m)
LayerControl().add_to(m)

m.save(os.path.join('results', 'FeatureGroup.html'))

m
```
